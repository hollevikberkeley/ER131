```python
import os
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import json
from shapely.geometry import Polygon
import folium

import warnings
warnings.filterwarnings('ignore')
```

# File Processing

```python
path = 'emb_data/SF_OAK_embed/focus_area/'
files = os.listdir(path)
files
```

```python
all_data = list()
for file in files:
    with open(path + file,'r') as f:
        data = json.loads(f.read())
        all_data.append(data)
```

```python
df = pd.json_normalize(all_data, record_path=['features'])
df
```

```python
df.to_csv('/Users/peter-evoy/Desktop/ER131_proj/processed_data/embeddings.csv')
```

# Plot to Map

```python
lists = df.loc[:,['geometry.coordinates']].iloc[:,0]
lists[0]
```

```python
poly = []
for i in range(0, len(lists)):
    poly.append(Polygon(lists[i][0]))
```

```python
poly_df = gpd.GeoDataFrame(geometry=poly, crs='epsg:4326')
poly_df
```

```python
sample = poly_df.sample(5000)
```

```python
m = folium.Map([37.8272, -122.2913], zoom_start=10, tiles='cartodbpositron')
folium.GeoJson(sample).add_to(m)
folium.LatLngPopup().add_to(m)
m
```
