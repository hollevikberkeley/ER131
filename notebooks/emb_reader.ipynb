```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import json
from shapely.geometry import Polygon
import folium

import warnings
warnings.filterwarnings('ignore')
```

```python
with open('emb_data/10SDH_32610_32_16_10_all.json','r') as f:
    data = json.loads(f.read())
    
df = pd.json_normalize(data, record_path=['features'])
```

```python
df.head()
```

```python
lists = df.loc[:,['geometry.coordinates']].iloc[:,0]
lists[0]
```

```python
poly = []
for i in range(0, len(lists)):
    poly.append(Polygon(lists[i][0]))
```

```python
poly_df = gpd.GeoDataFrame(geometry=poly, crs='epsg:4326')
poly_df
```

```python
poly_sub = poly_df.sample(5000)
poly_sub
```

```python
m = folium.Map([122.2913, 37.8272], zoom_start=5, tiles='cartodbpositron')
folium.GeoJson(poly_sub).add_to(m)
folium.LatLngPopup().add_to(m)
m
```
